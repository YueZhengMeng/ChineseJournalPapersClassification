```python
import pandas as pd
import csv
data396k = pd.read_table('../data/data40k.tsv', sep='\t', encoding='UTF-8', header=0, on_bad_lines='skip',quoting=csv.QUOTE_NONE)
data396k.insert(loc=0, column='sentence', value='')
data396k
```

```python
from tqdm import tqdm
for i in tqdm(range(data396k.shape[0])):
    sentence = data396k['title'][i] + "[PAD]"
    keywords = data396k['keywords'][i].split('_')
    for j in range(len(keywords)):
        sentence = sentence + keywords[j] + '[PAD]'
    sentence = sentence + data396k['abstract'][i]
    data396k['sentence'][i] = sentence
data396k
```

```python
data396k = data396k.drop(labels='title', axis=1)
data396k = data396k.drop(labels='abstract', axis=1)
data396k = data396k.drop(labels='keywords', axis=1)
data396k = data396k.drop(labels='discipline', axis=1)
data396k = data396k.drop(labels='category', axis=1)
data396k.insert(loc=1, column='length', value=0)
data396k
```

```python
from transformers import BertTokenizer
model_path = 'bert-base-chinese'
vocab_path = model_path
tokenizer = BertTokenizer.from_pretrained(vocab_path)
```

```python
data396k['sentence'][0]
```

```python
encode = tokenizer.encode(data396k['sentence'][0])
len(encode)
```

```python
#这一块代码运行需要约十分钟
from tqdm import tqdm
for i in tqdm(range(data396k.shape[0])):
        data396k['length'][i] = len(tokenizer.encode(data396k['sentence'][i]))

data396k
```

```python
s = 0
l = [0] * 100
for i in tqdm(range(data396k.shape[0])):
        t = data396k['length'][i]
        s = s + t
        if t//10 > 99:
                l[99] = l[99] + 1
        else:
                l[t//10] = l[t//10]  + 1
```

```python
#平均句长
s/data396k.shape[0]
```
