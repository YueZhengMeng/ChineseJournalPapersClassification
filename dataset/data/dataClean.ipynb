```python
import pandas as pd
import csv
csl = pd.read_table('./csl40k.tsv', sep='\t', encoding='UTF-8', header=0, on_bad_lines='skip',quoting=csv.QUOTE_NONE)
csl
```

```python
len(csl)
```

```python
from tqdm import tqdm
raws = set()
#标记NAN，NoneType或其他数据类型有错误的数据
for i in tqdm(range(len(csl))):
    if type('abc') != type(csl['category'][i]) or type('abc') != type(csl['discipline'][i]) or type('abc') != type(
            csl['keywords'][i]) or type('abc') != type(csl['abstract'][i]) or type('abc') != type(csl['title'][i]):
        raws.add(i)
```

```python
len(raws)
```

```python
#清除乱码
for i in tqdm(range(len(csl))):
    if i in raws:
        continue
    if '@@' in csl['title'][i]:
        csl['title'][i] = csl['title'][i].replace("@@","")
    if '@@' in csl['abstract'][i]:
        csl['abstract'][i] = csl['abstract'][i].replace("@@","")
    if '@@' in csl['keywords'][i]:
        csl['keywords'][i] = csl['keywords'][i].replace("@@","")
    if '@@' in csl['discipline'][i]:
        csl['discipline'][i] = csl['discipline'][i].replace("@@","")
    if '@@' in csl['category'][i]:
        csl['category'][i] = csl['category'][i].replace("@@","")
```

```python
#标记有明显残缺的数据
for i in tqdm(range(len(csl))):
    if i in raws:
        continue
    if len(csl['title'][i]) < 2 or len(csl['abstract'][i]) < 2 or len(csl['keywords'][i]) < 2 or len(
            csl['discipline'][i]) < 2 or len(csl['category'][i]) < 2:
        raws.add(i)
```

```python
len(raws)
```

```python
csl = csl.drop(raws, axis=0)
```

```python
len(csl)
```

```python
csl.to_csv('./csl40k.tsv', index=False, header=True, sep='\t')
```
