```python
import pandas as pd
import csv
data396k = pd.read_table('../data/data396k.tsv', sep='\t', encoding='UTF-8', header=0, on_bad_lines='skip',quoting=csv.QUOTE_NONE)
data396k.insert(loc=0, column='sentence', value='')
data396k = data396k.drop(labels='category', axis=1)
data396k = data396k.drop(labels='discipline', axis=1)
data396k
```

```python
from tqdm import tqdm
for i in tqdm(range(data396k.shape[0])):
    sentence = data396k['title'][i] + "[PAD]"
    keywords = data396k['keywords'][i].split('_')
    for j in range(len(keywords)):
        sentence = sentence + keywords[j] + '[PAD]'
    sentence = sentence + data396k['abstract'][i]
    data396k['sentence'][i] = sentence
data396k = data396k.drop(labels='title', axis=1)
data396k = data396k.drop(labels='abstract', axis=1)
data396k = data396k.drop(labels='keywords', axis=1)
#data396k = data396k.sample(frac=1.0)  # 将数据打乱
data396k
```

```python
f = open('./longStn.txt','w',encoding='UTF-8')
f
```

```python
for i in tqdm(range(data396k.shape[0])):
    f.write(data396k['sentence'][i]+'\n')
```

```python
from tqdm import tqdm
s = 0
m =0
l = [0] * 100

f = open('longStn.txt','r',encoding='UTF-8')
total_data = f.readlines()
with tqdm(total_data) as loader:
    for data in loader:
        t = len(data)
        s = s + t
        if t//10 > 99:
                l[99] = l[99] + 1
        else:
                l[t//10] = l[t//10]  + 1
print(s//len(total_data))
print(l)
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(20,8),dpi=80)

# 条形图宽度小于刻度时 每个类别的条形图分开
plt.bar(range(len(l)),l, align='center',width=0.8)
plt.xticks(range(len(l)), [i * 10 for i in range(len(l))],rotation=90)

plt.grid(alpha=0.4)
plt.tight_layout()
plt.savefig('LongStn_result_plot.png')
plt.show()
```
