```python
import pandas as pd
import csv
data396k = pd.read_table('../data/data396k.tsv', sep='\t', encoding='UTF-8', header=0, on_bad_lines='skip',quoting=csv.QUOTE_NONE)
data396k
```

```python
f = open('shortStn.txt','w',encoding='UTF-8')
f
```

```python
from tqdm import tqdm
from ltp import StnSplit
for i in tqdm(range(data396k.shape[0])):
    titleStn = StnSplit().split(data396k['title'][i])
    for t in titleStn:
        f.write(t+"\n")
    keywords = data396k['keywords'][i].split('_')
    k=''
    for j in range(len(keywords)):
        if j == len(keywords)-1:
            k = k +keywords[j]+"\n"
        else:
            k = k + keywords[j] + '[PAD]'
    f.write(k)
    abstractStn = StnSplit().split(data396k['abstract'][i])
    for a in abstractStn:
        f.write(a+"\n")
```

```python
f.close()
```

```python
s = 0
m =0
l = [0] * 30

f = open('shortStn.txt','r',encoding='UTF-8')
total_data = f.readlines()
with tqdm(total_data) as loader:
    for data in loader:
        t = len(data)
        s = s + t
        if t//10 > 29:
                l[29] = l[29] + 1
        else:
                l[t//10] = l[t//10]  + 1
print(s//len(total_data))
print(l)
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(20,8),dpi=80)

# 条形图宽度小于刻度时 每个类别的条形图分开
plt.bar(range(len(l)),l, align='center',width=0.8)
plt.xticks(range(len(l)), [i * 10 for i in range(len(l))],rotation=90)

plt.grid(alpha=0.4)
plt.tight_layout()
plt.savefig('shortStn_result_plot.png')
plt.show()
```
